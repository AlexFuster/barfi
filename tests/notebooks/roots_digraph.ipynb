# Find roots of a grph

```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
d = {1:['a1', 'a2', 'a3'], 2:['a4', 'a5','a7']}
G = nx.from_dict_of_lists(d, create_using=nx.DiGraph)
print(G.edges())
```

```python
plt.subplots(figsize=(12,6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=500)
```

To find the root nodes in each component, we first need to generate the induced subgraphs from the existent connected components, for that we have Graph.subgraph. Then, on each subgraph, the root node can be found by searching over all nodes, and keeping the one with an in_degree of 0.

```python
roots = []
for component in nx.weakly_connected_components(G):
    G_sub = G.subgraph(component)
    roots.extend([n for n,d in G_sub.in_degree() if d==0])
```

```python
print(roots)
```
